# Core
> Core functionality for the fastai audio library.

In this module we have the core functionality that enables you to:

* List all of the audio files in a folder (optionally recursive)
* Read audio files into torch tensors
* Transform the audio signals to other representations such as Spectrograms and MFCC's
* Configurations optimized to extract the best representation from different audio domains

To start using it, you only need one import:

```python
from fastai2_audio.core.all import *
```

And to download a sample dataset and list it's audio files, it's as simple as three lines of code:

```python
from fastai.data.external import *
data_path = untar_data(URLs.SPEAKERS10, extract_func=tar_extract_at_filename)
audio_files = get_audio_files(data_path)
```

You can easily create a torch tensor from an audio file, and this special tensor will have semantics to describe it's properties and how to plot/hear itself.

```python
audio = AudioTensor.create(audio_files[0])
print(f"""This audio have the following properties:
    sampling rate = {audio.sr}
    channels = {audio.nchannels}
    samples = {audio.nsamples}
""")
audio.show()
```
